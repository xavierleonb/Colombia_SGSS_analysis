```python
#Now, I´m gonna work with the departments and the ciphers of the entire population per each department. I´m working with estimations provided by Dane (Departamento Nacional de Estadísticas)

#It´s necessary standarise all names, especially for those ones who holds tildes and commas.
def clean_department_names(df):
    df["department"] = (df["department"]
                        .str.normalize("NFKD")
                        .str.encode("ascii", errors="ignore")
                        .str.decode("utf-8")
                        .str.upper()
                        .str.strip()
                        .str.replace("BOGOTA, D.C.", "BOGOTA D.C.")
                        .str.replace("ARCHIPIELAGO DE SAN ANDRES", "SAN ANDRES"))
    return df

summary_by_region = clean_department_names(summary_by_region)
estimates_population_df = clean_department_names(estimates_population_df)
estimates_population_df["quantity_habitants"] = pd.to_numeric(estimates_population_df["quantity_habitants"])

estimates_population_df = estimates_population_df[estimates_population_df["department"] != "NACIONAL"]

df_population_users = pd.merge(
    summary_by_region,
    estimates_population_df,
    on="department",
    how="inner"
)


df_population_users["total_affiliates"] = df_population_users["COTIZANTE"] + df_population_users["BENEFICIARIO"]
df_population_users["coverage_rate"] = df_population_users["total_affiliates"] / df_population_users["quantity_habitants"]


print("--- Coverage Rate per Department ---")
print(df_population_users.sort_values(by="coverage_rate", ascending=False))
```
