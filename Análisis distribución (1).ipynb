```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

df_colombiaSGSSS = pd.read_csv(r"C:\Users\GAZPRO\Desktop\Análisis de datos\ANÁLISIS SGSSS\data_SGSSS_new.csv")

df_colombiaSGSSS.head(20)
```

```python
print(df_colombiaSGSSS[df_colombiaSGSSS["entity_name"].str.contains("-CM", na=False)].head())
```

```python
df_colombiaSGSSS["entity_name"] = df_colombiaSGSSS["entity_name"].str.replace("-CM", "")
df_colombiaSGSSS["entity_name"] = df_colombiaSGSSS["entity_name"].str.strip()
print(df_colombiaSGSSS)
```

```python
total_affiliates_country = df_colombiaSGSSS["total_quantity"].sum()
print(total_affiliates_country)
```

```python
city_totals = df_colombiaSGSSS.groupby("municipality")["total_quantity"].sum()
top_ten_cities = city_totals.sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_ten_cities.values, y=top_ten_cities.index, palette="viridis")

plt.title("Top 10 municipios por cantidad de afiliados", fontsize=16)
plt.xlabel("Cantidad total de afiliados (en millones)", fontsize=12)
plt.ylabel("Municipio", fontsize=12)

for index, value in enumerate(top_ten_cities):
    plt.text(value, index, f" {value/1000000:.1f}M", va= "center")

plt.show()
```

```python
total_affiliates_top_3 = top_ten_cities.head(3).sum()

percentage = (total_affiliates_top_3 / total_affiliates_country) *100
print(f"El procentaje en las 3 grandes ciudades es: {percentage:.2f}%")
```
